**Features of** **datasets** 

1.age - age in years

2.sex - (1 = male ; 0 = female)

3.cp - chest pain type = 0: Typical angina : chest pain related decrease blood type

1 : Atypical angina 2: Non-anginal pain 3 : Asymoptomatic

4.trestbps - resting blood pressure

5.chol - serum cholestrol in mg/DL

6.fbs - fasting blood sugar

7.restecg - resting ecg results 0: normal 1: having ST-T wave abnormality (T wave inversion and/or ST elevation or depression of > 0.05 mV) 2. showing probable or definite left ventricular hypertrophy by Estes'criteria

8.thalach - maximum heart rate achieved

9.exang - exercise-include angina (True / False)

10.oldpeak - ST depression induced by exercise relative to rest

11.slope - the slope of the peak exercise ST segment

12.ca - number of major vessels (0-3) colored by fluoroscopy

13.thal - normal; fixed defect; reversible defect (thalasemia indicator)

14.the predicted attribute
